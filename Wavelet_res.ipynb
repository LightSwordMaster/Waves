# –õ–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–∞—è 1 –í–≤–µ–¥–µ–Ω–∏–µ –≤ –≤–µ–π–≤–ª–µ—Ç-–∞–Ω–∞–ª–∏–∑

–¶–µ–ª—å —Ä–∞–±–æ—Ç—ã:
 - –ü–æ–ª—É—á–∏—Ç—å –æ–±—â–µ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–∏ –æ –≤–µ–π–≤–ª–µ—Ç-–∞–Ω–∞–ª–∏–∑–µ –∏ –æ—Å–Ω–æ–≤–Ω—ã—Ö —Ç–∏–ø–∞—Ö –≤–µ–π–≤–ª–µ—Ç–æ–≤, –ø–æ–Ω—è—Ç—å –æ—Å–Ω–æ–≤–Ω—ã–µ —Å–≤–π—Å—Ç–≤–∞ –∏ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤–µ–π–≤–ª–µ—Ç–æ–≤.

–ê–Ω–≥–ª–∏–π—Å–∫–æ–µ —Å–ª–æ–≤–æ wavelet (–æ—Ç —Ñ—Ä–∞–Ω—Ü—É–∑—Å–∫–æ–≥–æ ¬´ondelette¬ª) –¥–æ—Å–ª–æ–≤–Ω–æ –ø–µ—Ä–µ–≤–æ–¥–∏—Ç—Å—è –∫–∞–∫ ¬´–∫–æ—Ä–æ—Ç–∫–∞—è (–º–∞–ª–µ–Ω—å–∫–∞—è) –≤–æ–ª–Ω–∞¬ª. –í —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –ø–µ—Ä–µ–≤–æ–¥–∞—Ö –∑–∞—Ä—É–±–µ–∂–Ω—ã—Ö —Å—Ç–∞—Ç–µ–π –Ω–∞ —Ä—É—Å—Å–∫–∏–π —è–∑—ã–∫ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –µ—â–µ —Ç–µ—Ä–º–∏–Ω—ã: ¬´–≤—Å–ø–ª–µ—Å–∫¬ª, ¬´–≤—Å–ø–ª–µ—Å–∫–æ–≤–∞—è —Ñ—É–Ω–∫—Ü–∏—è¬ª, ¬´–º–∞–ª–æ–≤–æ–ª–Ω–æ–≤–∞—è —Ñ—É–Ω–∫—Ü–∏—è¬ª, ¬´–≤–æ–ª–Ω–æ—á–∫–∞¬ª –∏ –¥—Ä.  

–í–µ–π–≤–ª–µ—Ç-–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ (–í–ü) —à–∏—Ä–æ–∫–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ —Å–∏–≥–Ω–∞–ª–æ–≤. –ü–æ–º–∏–º–æ —ç—Ç–æ–≥–æ, –æ–Ω–æ –Ω–∞—Ö–æ–¥–∏—Ç –±–æ–ª—å—à–æ–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –≤ –æ–±–ª–∞—Å—Ç–∏ —Å–∂–∞—Ç–∏—è –¥–∞–Ω–Ω—ã—Ö. –í–ü –æ–¥–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Å–∏–≥–Ω–∞–ª–∞ ‚Äì —ç—Ç–æ –µ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –≤–≤–∏–¥–µ –æ–±–æ–±—â–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞ –∏–ª–∏ –∏–Ω—Ç–µ–≥—Ä–∞–ª–∞ –§—É—Ä—å–µ –ø–æ —Å–∏—Å—Ç–µ–º–µ –±–∞–∑–∏—Å–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π:



—Å–∫–æ–Ω—Å—Ç—Ä—É–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∏–∑ –º–∞—Ç–µ—Ä–∏–Ω—Å–∫–æ–≥–æ (–∏—Å—Ö–æ–¥–Ω–æ–≥–æ) –≤–µ–π–≤–ª–µ—Ç–∞ , –æ–±–ª–∞–¥–∞—é—â–µ–≥–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–º–∏ —Å–≤–æ–π—Å—Ç–≤–∞–º–∏ –∑–∞ —Å—á–µ—Ç –æ–ø–µ—Ä–∞—Ü–∏–π —Å–¥–≤–∏–≥–∞ –≤–æ –≤—Ä–µ–º–µ–Ω–∏ ( b ) –∏ –∏–∑–º–µ–Ω–µ–Ω–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –º–∞—Å—à—Ç–∞–±–∞ (a).

–ú–Ω–æ–∂–∏—Ç–µ–ª—å $  \frac{1}{\sqrt{a}} $ –æ–±–µ—Å–ø–µ—á–∏–≤–∞–µ—Ç –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –Ω–æ—Ä–º—ã —Ñ—É–Ω–∫—Ü–∏–π (1) –æ—Ç –º–∞—Å—à—Ç–∞–±–∏—Ä—É—é—â–µ–≥–æ —á–∏—Å–ª–∞ (a). –î–ª—è –∑–∞–¥–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ a –∏ b —Ñ—É–Ω–∫—Ü–∏—è $ ùúì_{ab} (t) $ –∏ –µ—Å—Ç—å –≤–µ–π–≤–ª–µ—Ç, –ø–æ—Ä–æ–∂–¥–∞–µ–º—ã–π –º–∞—Ç–µ—Ä–∏–Ω—Å–∫–∏–º –≤–µ–π–≤–ª–µ—Ç–æ–º $ ùúì(t)$.

–û–¥–Ω–æ  –∏–∑ –æ—Å–Ω–æ–≤–Ω—ã—Ö –æ—Ç–ª–∏—á–∏–π –≤–µ–π–≤–ª–µ—Ç–æ–≤-–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –æ—Ç –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –§—É—Ä—å–µ - –ª–æ–∫–∞–ª—å–Ω–æ—Å—Ç—å –≤–æ –≤—Ä–µ–º–µ–Ω–∏, —á—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω—ã–µ —Å–∏–≥–Ω–∞–ª—ã. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –§—É—Ä—å–µ –±—É–¥–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å –æ—á–µ–Ω—å —Ö–æ—Ä–æ—à–æ, –∫–æ–≥–¥–∞ —á–∞—Å—Ç–æ—Ç–Ω—ã–π —Å–ø–µ–∫—Ç—Ä —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω—ã–π. –ü—Ä–∏ —ç—Ç–æ–º —á–∞—Å—Ç–æ—Ç—ã, –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—â–∏–µ –≤ —Å–∏–≥–Ω–∞–ª–µ, –Ω–µ –∑–∞–≤–∏—Å—è—Ç –æ—Ç –≤—Ä–µ–º–µ–Ω–∏, –∏ —Å–∏–≥–Ω–∞–ª —Å–æ–¥–µ—Ä–∂–∏—Ç —á–∞—Å—Ç–æ—Ç—ã xHz, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É–µ—Ç –≤ –ª—é–±–æ–º –º–µ—Å—Ç–µ —Å–∏–≥–Ω–∞–ª–∞. –ß–µ–º –Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–µ–µ —Å–∏–≥–Ω–∞–ª, —Ç–µ–º —Ö—É–∂–µ –±—É–¥—É—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã. –≠—Ç–æ –ø—Ä–æ–±–ª–µ–º–∞, —Ç–∞–∫ –∫–∞–∫ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ —Å–∏–≥–Ω–∞–ª–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –º—ã –≤–∏–¥–∏–º –≤ —Ä–µ–∞–ª—å–Ω–æ–π –∂–∏–∑–Ω–∏, –Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω—ã –ø–æ —Å–≤–æ–µ–π –ø—Ä–∏—Ä–æ–¥–µ. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –§—É—Ä—å–µ –∏–º–µ–µ—Ç –≤—ã—Å–æ–∫–æ–µ —Ä–∞–∑—Ä–µ—à–µ–Ω–∏–µ –≤ —á–∞—Å—Ç–æ—Ç–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏, –Ω–æ –Ω—É–ª–µ–≤–æ–µ —Ä–∞–∑—Ä–µ—à–µ–Ω–∏–µ –≤–æ –≤—Ä–µ–º–µ–Ω–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏. 

–ü–æ–∫–∞–∂–µ–º —ç—Ç–æ –Ω–∞ –ø—Ä–∏–º–µ—Ä–µ:

```python
import numpy as np
from scipy import fftpack
from pylab import*
N=100000
dt = 1e-5
xa = np.linspace(0, 1, num=N)
xb = np.linspace(0, 1/4, num=N/4) 
frequencies = [4, 30, 60, 90]
y1a, y1b = np.sin(2*np.pi*frequencies[0]*xa), np.sin(2*np.pi*frequencies[0]*xb)
y2a, y2b = np.sin(2*np.pi*frequencies[1]*xa), np.sin(2*np.pi*frequencies[1]*xb)
y3a, y3b = np.sin(2*np.pi*frequencies[2]*xa), np.sin(2*np.pi*frequencies[2]*xb)
y4a, y4b = np.sin(2*np.pi*frequencies[3]*xa), np.sin(2*np.pi*frequencies[3]*xb)
def spectrum_w(y):    
    Fs = 1 / dt  #  Fs = 0,1 MHz 
    n = len(y)  # –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–∏–≥–Ω–∞–ª–∞
    k = np.arange(n)
    T = n / Fs
    frq = k / T  # —á–∞—Å—Ç–æ—Ç–Ω—ã–π –¥–∏–∞–ø–∞–∑–æ–Ω x2
    frq = frq[range(n // 2)]  # one side frequency range
    Y = fftpack.fft(y) / n 
    Y = Y[range(n // 2)] / max(Y[range(n // 2)])
    subplot(2, 1, 1)
    plot(k/N , y, 'b')    
    ylabel('Amplitude')
    grid()
    subplot(2, 1, 2)
    plot(frq[0:140], abs(Y[0:140]), 'r')
    xlabel('Freq')
    plt.ylabel('|Y(freq)|')
    grid() 
y= y1a + y2a + y3a + y4a
y1 = np.concatenate([y1b, y2b, y3b, y4b])
print("–°—É–º–º–∞ —á–µ—Ç—ã—Ä—ë—Ö –≥–∞—Ä–º–æ–Ω–∏–∫, –ø–æ–ª–∏–≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–∏–π —Å–∏–≥–Ω–∞–ª")
spectrum_w(y)
show()
print("–°–∏–≥–Ω–∞–ª —Å –∏–∑–º–µ–Ω—è—é—â–µ–π—Å—è —á–∞—Å—Ç–æ—Ç–æ–π, 4 —Ä–∞–∑–Ω—ã–µ –≥–∞—Ä–º–æ–Ω–∏–∫–∏ —Å–ª–µ–¥—É—é—Ç –¥—Ä—É–≥ –∑–∞ –¥—Ä—É–≥–æ–º")
spectrum_w(y1)
show()
```

–ì–ª–∞–≤–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤–µ–π–≤–ª–µ—Ç–∞

**–û–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å**. –ö–≤–∞–¥—Ä–∞—Ç –Ω–æ—Ä–º—ã —Ñ—É–Ω–∫—Ü–∏–∏ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –∫–æ–Ω–µ—á–Ω—ã–º:
$\left \| \psi \right \|^{2}=\int_{-\infty }^{\infty }\left | \psi (t) \right |^{2}dt< \infty$. (2)

**–õ–æ–∫–∞–ª–∏–∑–∞—Ü–∏—è**. –í–ü –≤ –æ—Ç–ª–∏—á–∏–µ –æ—Ç –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –§—É—Ä—å–µ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –ª–æ–∫–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—É—é –∏—Å—Ö–æ–¥–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é –∏ –≤–æ –≤—Ä–µ–º–µ–Ω–∏, –∏ –ø–æ —á–∞—Å—Ç–æ—Ç–µ. –î–ª—è —ç—Ç–æ–≥–æ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ, —á—Ç–æ–±—ã –≤—ã–ø–æ–ª–Ω—è–ª–∏—Å—å —É—Å–ª–æ–≤–∏—è:

 $\left | \psi (t) \right |\leq C(1+\left | t \right |)^{-1-\varepsilon } $ –∏
 
 $\left | S_{\psi }(\omega ) \right |\leq C(1+\left | \omega \right |)^{-1-\varepsilon } $ –ø—Ä–∏ $\varepsilon > 0$, (3)

–ù–∞–ø—Ä–∏–º–µ—Ä, –¥–µ–ª—å—Ç–∞-—Ñ—É–Ω–∫—Ü–∏—è $\delta(t)$ –∏ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è –Ω–µ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–º—É —É—Å–ª–æ–≤–∏—é –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ–π –ª–æ–∫–∞–ª–∏–∑–∞—Ü–∏–∏ –≤–æ –≤—Ä–µ–º–µ–Ω–Ω–æ–π –∏ —á–∞—Å—Ç–æ—Ç–Ω–æ–π –æ–±–ª–∞—Å—Ç—è—Ö.

**–ù—É–ª–µ–≤–æ–µ —Å—Ä–µ–¥–Ω–µ–µ.** –ì—Ä–∞—Ñ–∏–∫ –∏—Å—Ö–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –¥–æ–ª–∂–µ–Ω –æ—Å—Ü–∏–ª–ª–∏—Ä–æ–≤–∞—Ç—å (–±—ã—Ç—å –∑–Ω–∞–∫–æ–ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º) –≤–æ–∫—Ä—É–≥ –Ω—É–ª—è –Ω–∞ –æ—Å–∏ –≤—Ä–µ–º–µ–Ω–∏ –∏ –∏–º–µ—Ç—å –Ω—É–ª–µ–≤—É—é –ø–ª–æ—â–∞–¥—å:

$\int_{-\infty }^{\infty }\psi (t)dt=0$. (4)

–ò–∑ —ç—Ç–æ–≥–æ —É—Å–ª–æ–≤–∏—è —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –ø–æ–Ω—è—Ç–Ω—ã–º –≤—ã–±–æ—Ä –Ω–∞–∑–≤–∞–Ω–∏—è ¬´–≤–µ–π–≤–ª–µ—Ç¬ª ‚Äì –º–∞–ª–µ–Ω—å–∫–∞—è –≤–æ–ª–Ω–∞.

–†–∞–≤–µ–Ω—Å—Ç–≤–æ –Ω—É–ª—é –ø–ª–æ—â–∞–¥–∏ —Ñ—É–Ω–∫—Ü–∏–∏ $\psi(t)$, —Ç.–µ. –Ω—É–ª–µ–≤–æ–≥–æ –º–æ–º–µ–Ω—Ç–∞, –ø—Ä–∏–≤–æ–¥–∏—Ç –∫ —Ç–æ–º—É, —á—Ç–æ —Ñ—É—Ä—å–µ-–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ $S_{\psi}(\omega)$ —ç—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞–≤–Ω–æ –Ω—É–ª—é –ø—Ä–∏ $\omega$ –∏ –∏–º–µ–µ—Ç –≤–∏–¥ –ø–æ–ª–æ—Å–æ–≤–æ–≥–æ —Ñ–∏–ª—å—Ç—Ä–∞. –ü—Ä–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö (a) —ç—Ç–æ –±—É–¥–µ—Ç –Ω–∞–±–æ—Ä –ø–æ–ª–æ—Å–æ–≤—ã—Ö —Ñ–∏–ª—å—Ç—Ä–æ–≤.

–ß–∞—Å—Ç–æ –¥–ª—è –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–π –±—ã–≤–∞–µ—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ, —á—Ç–æ–±—ã –Ω–µ —Ç–æ–ª—å–∫–æ –Ω—É–ª–µ–≤–æ–π, –Ω–æ –∏ –≤—Å–µ –ø–µ—Ä–≤—ã–µ n –º–æ–º–µ–Ω—Ç–æ–≤ –±—ã–ª–∏ —Ä–∞–≤–Ω—ã –Ω—É–ª—é:

$\int_{-\infty }^{\infty }t^{^{n}}\psi (t)dt=0$. (5)

–í–µ–π–≤–ª–µ—Ç—ã n -–≥–æ –ø–æ—Ä—è–¥–∫–∞ –ø–æ–∑–≤–æ–ª—è—é—Ç –∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –±–æ–ª–µ–µ —Ç–æ–Ω–∫—É—é (–≤—ã—Å–æ–∫–æ—á–∞—Å—Ç–æ—Ç–Ω—É—é) —Å—Ç—Ä—É–∫—Ç—É—Ä—É —Å–∏–≥–Ω–∞–ª–∞, –ø–æ–¥–∞–≤–ª—è—è –º–µ–¥–ª–µ–Ω–Ω–æ –∏–∑–º–µ–Ω—è—é—â–∏–µ—Å—è –µ–≥–æ —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏–µ.

**–ê–≤—Ç–æ–º–æ–¥–µ–ª—å–Ω–æ—Å—Ç—å.** –•–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã–º –ø—Ä–∏–∑–Ω–∞–∫–æ–º –í–ü —è–≤–ª—è–µ—Ç—Å—è –µ–≥–æ —Å–∞–º–æ–ø–æ–¥–æ–±–∏–µ. –í—Å–µ –≤–µ–π–≤–ª–µ—Ç—ã –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–≥–æ —Å–µ–º–µ–π—Å—Ç–≤–∞ $\psi_{ab}(t)$ –∏–º–µ—é—Ç —Ç–æ –∂–µ —á–∏—Å–ª–æ –æ—Å—Ü–∏–ª–ª—è—Ü–∏–π, —á—Ç–æ –∏ –º–∞—Ç–µ—Ä–∏–Ω—Å–∫–∏–π –≤–µ–π–≤–ª–µ—Ç $\psi(t)$, –ø–æ—Å–∫–æ–ª—å–∫—É –ø–æ–ª—É—á–µ–Ω—ã –∏–∑ –Ω–µ–≥–æ –ø–æ—Å—Ä–µ–¥—Å—Ç–≤–æ–º –º–∞—Å—à—Ç–∞–±–Ω—ã—Ö –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π ( a ) –∏ —Å–¥–≤–∏–≥–∞ ( b ).



–ü–µ—Ä–µ–π–¥—ë–º –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ –∫ —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–∏—é —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –≤–µ–π–≤–ª–µ—Ç —Ñ—É–Ω–∫—Ü–∏–π –∏ –≤–ª–∏—è–Ω–∏—é –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Å–¥–≤–∏–≥–∞ –≤–æ –≤—Ä–µ–º–µ–Ω–∏ b –º –∏–∑–º–µ–Ω–µ–Ω–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –º–∞—Å—à—Ç–∞–±–∞ –Ω–∞ –∏—Ö –≤–∏–¥.

```python
import numpy as np
import pywt
import matplotlib.pyplot as plt

def on_button_clicked(b):
    vav='gaus{}'.format(slider_g1.value)
    wav = pywt.ContinuousWavelet(vav)
    # –≤—ã–≤–µ—Å—Ç–∏ –¥–∏–∞–ø–∞–∑–æ–Ω, –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥–µ—Ç –æ—Ü–µ–Ω–∏–≤–∞—Ç—å—Å—è –≤–µ–π–≤–ª–µ—Ç
    print("–ù–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–π –≤–µ–π–≤–ª–µ—Ç –±—É–¥–µ—Ç –æ—Ü–µ–Ω–∏–≤–∞—Ç—å—Å—è –≤–æ –≤—Å–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω–µ [{}, {}]".format(   wav.lower_bound, wav.upper_bound))
    width = wav.upper_bound - wav.lower_bound
    scales = [1, 2, 3, 4, 10, 15]
    max_len = int(np.max(scales)*width + 1)
    t = np.arange(max_len)
    fig, axes = plt.subplots(len(scales), 2, figsize=(12, 6))
    for n, scale in enumerate(scales):
        # –°–ª–µ–¥—É—é—â–∏–π –∫–æ–¥ –∞–¥–∞–ø—Ç–∏—Ä–æ–≤–∞–Ω –∏–∑ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏—Ö —á–∞—Å—Ç–µ–π cwt
        int_psi, x = pywt.integrate_wavelet(wav, precision=10)
        step = x[1] - x[0]
        j = np.floor(
            np.arange(scale * width + 1) / (scale * step))
        if np.max(j) >= np.size(int_psi):
            j = np.delete(j, np.where((j >= np.size(int_psi)))[0])
        j = j.astype(np.int)
        # normalize int_psi –¥–ª—è –±–æ–ª–µ–µ –ø—Ä–æ—Å—Ç–æ–≥–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è
        int_psi /= np.abs(int_psi).max()
      # –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–µ –≤—ã–±–æ—Ä–∫–∏ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –≤–µ–π–≤–ª–µ—Ç–∞
        filt = int_psi[j][::-1]
    # CWT —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ —Å–≤–µ—Ä—Ç–∫–∏ —Ñ–∏–ª—å—Ç—Ä–∞ —Å —Å–∏–≥–Ω–∞–ª–æ–º –≤ —ç—Ç–æ–π —à–∫–∞–ª–µ
    # –ó–¥–µ—Å—å –º—ã —Å—Ç—Ä–æ–∏–º —ç—Ç–æ –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–µ —è–¥—Ä–æ —Å–≤–µ—Ä—Ç–∫–∏ –≤ –∫–∞–∂–¥–æ–º –º–∞—Å—à—Ç–∞–±–µ.
        nt = len(filt)
        t = np.linspace(-nt//2, nt//2, nt)
        axes[n, 0].plot(t, filt.real, t, filt.imag)
        axes[n, 0].set_xlim([-max_len//2, max_len//2])
        axes[n, 0].set_ylim([-1, 1])
        axes[n, 0].text(50, 0.35, 'scale = {}'.format(scale))
        f = np.linspace(-np.pi, np.pi, max_len)
        filt_fft = np.fft.fftshift(np.fft.fft(filt, n=max_len))
        filt_fft /= np.abs(filt_fft).max()
        axes[n, 1].plot(f, np.abs(filt_fft)**2)
        axes[n, 1].set_xlim([-np.pi, np.pi])
        axes[n, 1].set_ylim([0, 1])
        axes[n, 1].set_xticks([-np.pi, 0, np.pi])
        axes[n, 1].set_xticklabels([r'$-\pi$', '0', r'$\pi$'])
        axes[n, 1].grid(True, axis='x')
        axes[n, 1].text(np.pi/2, 0.5, 'scale = {}'.format(scale))
    axes[n, 0].set_xlabel('time (samples)')
    axes[n, 1].set_xlabel('frequency (radians)')
    axes[0, 0].legend(['real', 'imaginary'], loc='upper left')
    axes[0, 1].legend(['Power'], loc='upper left')
    axes[0, 0].set_title('filter: wavelet - %s'%vav)
    axes[0, 1].set_title(r'|FFT(filter)|$^2$')
    plt.show()
    
print("–í–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–π –≤–µ–π–≤–ª–µ—Ç –ì–∞—É—Å—Å–∞ –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞")    
slider_g1 = widgets.IntSlider(min=1, max=8, continuous_update=True, description='–ü–æ—Ä—è–¥–æ–∫')
display(slider_g1)
button = widgets.Button(description="–ü–æ—Å—Ç—Ä–æ–∏—Ç—å")
button.on_click(on_button_clicked)
display(button)
```

# –õ–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–∞—è 2 –í–µ–π–≤–ª–µ—Ç-–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∏ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ —Ç–∏–ø–æ–≤—ã—Ö —Å–∏–≥–Ω–∞–ª–æ–≤

**–ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–µ –≤–µ–π–≤–ª–µ—Ç-–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ**

–ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–µ (–∏–Ω—Ç–µ–≥—Ä–∞–ª—å–Ω–æ–µ) –≤–µ–π–≤–ª–µ—Ç-–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ (–ù–í–ü –∏–ª–∏ –°WT ‚Äì continuous wavelet transform). –°–∫–æ–Ω—Å—Ç—Ä—É–∏—Ä—É–µ–º –±–∞–∑–∏—Å $\psi_{ab}(t)$ —Å –ø–æ–º–æ—â—å—é –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã—Ö –º–∞—Å—à—Ç–∞–±–Ω—ã—Ö –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π (a) –∏ –ø–µ—Ä–µ–Ω–æ—Å–æ–≤ (b) –º–∞—Ç–µ—Ä–∏–Ω—Å–∫–æ–≥–æ –≤–µ–π–≤–ª–µ—Ç–∞ $\psi(t)$ —Å –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –±–∞–∑–∏—Å–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ a –∏ b –≤ —Ñ–æ—Ä–º—É–ª–µ (1).

–¢–æ–≥–¥–∞, –ø–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é –ø—Ä—è–º–æ–µ (–∞–Ω–∞–ª–∏–∑) –∏ –æ–±—Ä–∞—Ç–Ω–æ–µ (—Å–∏–Ω—Ç–µ–∑) H–í–ü (—Ç.–µ. –ü–ù–í–ü –∏ –û–ù–í–ü) —Å–∏–≥–Ω–∞–ª–∞ S(t) –∑–∞–ø–∏—à—É—Ç—Å—è —Ç–∞–∫:
$W_{s}(a,b)=(S(t),\psi _{ab}(t))=\frac{1}{\sqrt{a}}\int_{-\infty }^{\infty }S(t)\psi \left ( \frac{t-b}{a} \right )dt$, (6)

$S(t)=\frac{1}{C_{\psi }}\int_{-\infty }^{\infty }\int_{-\infty }^{\infty }W_{s}(a,b)\psi _{ab}(t)\frac{dadb}{a^{2}}$, (7)

–≥–¥–µ $C_{\psi}$‚Äì –Ω–æ—Ä–º–∏—Ä—É—é—â–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç,
$C_{\psi }=\int_{-\infty }^{\infty }\left | \psi (\omega ) \right |^{2}\left | \omega \right |^{-1}d\omega < \infty$

–≥–¥–µ: (‚Ä¢,‚Ä¢) ‚Äì —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Å–æ–º–Ω–æ–∂–∏—Ç–µ–ª–µ–π,
$\psi(\omega)$‚Äì —Ñ—É—Ä—å–µ-–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤–µ–π–≤–ª–µ—Ç–∞ $\psi$. –î–ª—è –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –≤–µ–π–≤–ª–µ—Ç–æ–≤ $C_{\psi}$ .

–ò–∑ (6) —Å–ª–µ–¥—É–µ—Ç, —á—Ç–æ –≤–µ–π–≤–ª–µ—Ç-—Å–ø–µ–∫—Ç—Ä $W_s(b,a)$ (wavelet spectrum, –∏–ª–∏ time-scale-spectrum ‚Äì –º–∞—Å—à—Ç–∞–±–Ω–æ-–≤—Ä–µ–º–µ–Ω–Ω–æ–π —Å–ø–µ–∫—Ç—Ä) –≤ –æ—Ç–ª–∏—á–∏–µ –æ—Ç —Ñ—É—Ä—å–µ-—Å–ø–µ–∫—Ç—Ä–∞ (single spectrum) —è–≤–ª—è–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π –¥–≤—É—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤: –ø–µ—Ä–≤—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç –∞ (–≤—Ä–µ–º–µ–Ω–Ω–æ–π –º–∞—Å—à—Ç–∞–±) –∞–Ω–∞–ª–æ–≥–∏—á–µ–Ω –ø–µ—Ä–∏–æ–¥—É –æ—Å—Ü–∏–ª–ª—è—Ü–∏–π, —Ç.–µ. –æ–±—Ä–∞—Ç–µ–Ω —á–∞—Å—Ç–æ—Ç–µ, –∞ –≤—Ç–æ—Ä–æ–π b‚Äì–∞–Ω–∞–ª–æ–≥–∏—á–µ–Ω —Å–º–µ—â–µ–Ω–∏—é —Å–∏–≥–Ω–∞–ª–∞ –ø–æ –æ—Å–∏ –≤—Ä–µ–º–µ–Ω–∏.

–°–ª–µ–¥—É–µ—Ç –æ—Ç–º–µ—Ç–∏—Ç—å, —á—Ç–æ $W_s(b,a_0)$ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É–µ—Ç –≤—Ä–µ–º–µ–Ω–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å (–ø—Ä–∏ $a=a_0$, —Ç–æ–≥–¥–∞ –∫–∞–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ $W_s(a,b_0)$ –º–æ–∂–Ω–æ –ø–æ—Å—Ç–∞–≤–∏—Ç—å –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —á–∞—Å—Ç–æ—Ç–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å (–ø—Ä–∏ $b=b_0$).

–ï—Å–ª–∏ –∏—Å—Å–ª–µ–¥—É–µ–º—ã–π —Å–∏–≥–Ω–∞–ª S(t) –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π –æ–¥–∏–Ω–æ—á–Ω—ã–π –∏–º–ø—É–ª—å—Å –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é $\tau_{u}$, —Å–æ—Å—Ä–µ–¥–æ—Ç–æ—á–µ–Ω–Ω—ã–π –≤ –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ $t=t_0$, —Ç–æ –µ–≥–æ –≤–µ–π–≤–ª–µ—Ç-—Å–ø–µ–∫—Ç—Ä –±—É–¥–µ—Ç –∏–º–µ—Ç—å –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏ —Å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏ $a=\tau_u, b=t_0$.

–°–ø–æ—Å–æ–±—ã –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è (–≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏)  $W_s(b,a)$ –º–æ–≥—É—Ç –±—ã—Ç—å —Ä–∞–∑–ª–∏—á–Ω—ã–º–∏. –°–ø–µ–∫—Ç—Ä $W_s(b,a)$ —è–≤–ª—è–µ—Ç—Å—è –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å—é –≤ —Ç—Ä–µ—Ö–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ. –û–¥–Ω–∞–∫–æ, —á–∞—Å—Ç–æ –≤–º–µ—Å—Ç–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—Ç –µ—ë –ø—Ä–æ–µ–∫—Ü–∏—é –Ω–∞ –ø–ª–æ—Å–∫–æ—Å—Ç—å ab —Å –∏–∑–æ—É—Ä–æ–≤–Ω—è–º–∏ (–∏–ª–∏ —Ñ–∏–≥—É—Ä–∞–º–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ü–≤–µ—Ç–æ–≤), –ø–æ–∑–≤–æ–ª—è—é—â–∏–º–∏ –ø—Ä–æ—Å–ª–µ–¥–∏—Ç—å –∏–∑–º–µ–Ω–µ–Ω–∏–µ –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç–∏ –∞–º–ø–ª–∏—Ç—É–¥ –í–ü –Ω–∞ —Ä–∞–∑–Ω—ã—Ö –º–∞—Å—à—Ç–∞–±–∞—Ö (–∞) –∏ –≤–æ –≤—Ä–µ–º–µ–Ω–∏ (b ).

–ö—Ä–æ–º–µ —Ç–æ–≥–æ, –∏–∑–æ–±—Ä–∞–∂–∞—é—Ç –∫–∞—Ä—Ç–∏–Ω—ã –ª–∏–Ω–∏–π –ª–æ–∫–∞–ª—å–Ω—ã—Ö —ç–∫—Å—Ç—Ä–µ–º—É–º–æ–≤ —ç—Ç–∏—Ö –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–µ–π, —Ç–∞–∫ –Ω–∞–∑—ã–≤–∞–µ–º—ã–π —Å–∫–µ–ª–µ—Ç–æ–Ω (sceleton), –∫–æ—Ç–æ—Ä—ã–π –≤—ã—è–≤–ª—è–µ—Ç —Å—Ç—Ä—É–∫—Ç—É—Ä—É –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º–æ–≥–æ —Å–∏–≥–Ω–∞–ª–∞.

    - –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ —Å–∏–≥–Ω–∞–ª–æ–≤ –∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –≤–µ–π–≤–ª–µ—Ç –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –¥–ª—è —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    - –∑–∞—Ç–µ–º –∏–∑–º–µ–Ω—è–µ–º –æ–¥–∏–Ω –∏–∑ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –≤–æ–∑–≤—Ä–∞—â–∞—è—è –æ—Å—Ç–∞–ª—å–Ω—ã–µ –∫ –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º. –û—Ü–µ–Ω–∏—Ç—å –≤–ª–∏—è–Ω–∏–µ –ø–µ—Ä–∏–æ–¥–∞ –¢, –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–∏—è –∏ –∑–Ω–∞—á–µ–Ω–∏—è N –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è
    - –¥–ª—è –ø–æ–ª–∏–≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–≥–æ —Å–∏–≥–Ω–∞–ª–∞ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ø—Ä–∏ —Ä–∞–∑–ª–∏—á–Ω–æ–º –∫–æ–ª–∏—á—á–µ—Å—Ç–≤–µ –≥–∞—Ä–º–æ–Ω–∏–∫ (2,3,4), –∑–∞—Ç–µ–º –¥–ª—è –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º —á–∏—Å–ª–µ –≥–∞—Ä–º–æ–Ω–∏–∫ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –∫–∞–∫  –≤–ª–∏—è–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–∏–æ–¥–∞ T –∏ –∑–Ω–∞—á–Ω–∏–µ N –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    - –¥–ª—è –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–æ–≥–æ –∏–º–ø—É–ª—å—Å–∞ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ø—Ä–∏ 3-—Ö —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∏ –∑–∞—Ç–µ–º –æ—Ü–µ–Ω–∏—Ç—å –≤–ª–∏—è–Ω–∏–µ N (–º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ, —Å—Ä–µ–¥–Ω–µ–µ, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ).
    - –∏–∑-–∑–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∞ –ø—Ä–µ–æ–±—Ä–∞–æ–≤–∞–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∏ —Å—Ç—Ä–æ—è—Ç—Å—è –¥–æ–ª–≥–æ, —Ç–∞–∫ —á—Ç–æ —Å—Ç–æ–∏—Ç –ø–æ–¥–æ–∂–¥–∞—Ç—å. –í –¥–∞–ª—å–Ω–µ–π—à–µ–º –±—É–¥–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –±—ã—Å—Ç—Ä–æ–µ –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–µ –≤–µ–π–≤–ª–µ—Ç-–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–ø–µ–∫—Ç—Ä–æ–≥—Ä–∞–º–º. –û–Ω–∏ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—é—Ç—Å—è –≤ —Å–ª–µ–¥—É—é—â–µ–π –ª–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–æ–π.

**–ì–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –∫–æ–ª–µ–±–∞–Ω–∏–µ** 

–°–∏–≥–Ω–∞–ª –∏–º–µ–µ—Ç –≤–∏–¥: $s(t)=Asin(\omega t-\phi)$

–≥–¥–µ: $A=1 V,\omega=\frac{2\pi }{T}=\frac{2\pi }{50},\psi=0$ 

–í–µ–π–≤–ª–µ—Ç–æ–±—Ä–∞–∑—É—é—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è: $MHAT(t):=\frac{d^{2}}{dt^{2}}exp(-t^{2}/2)$,

–í–µ–π–≤–ª–µ—Ç—ã: $\psi (a,b,t)=\frac{1}{\sqrt{a}}MHAT\left ( \frac{t-b}{a} \right )$,

–í–µ–π–≤–ª–µ—Ç-—Å–ø–µ–∫—Ç—Ä: N:=256, a:=1..30, b:=0..50,  $W(a,b):=\int_{-N}^{N}\psi (a,b,t)s(t)dt, N_{ab}:=W(a,b)$

```python
from scipy.integrate import quad
from numpy import*
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

def on_button_clicked(b):
    N=drop_N.value
    T=slider_T2.value
    T_s = T*drop_duration4.value
    N_t=round(T*drop_duration4.value/2)
    def S(t):
        return sin(2*pi*t/T)
    plt.figure()
    plt.title(' –ì–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –∫–æ–ª–µ–±–∞–Ω–∏–µ', size=12)
    y=[S(t) for t in arange(0,T_s,1)]
    x=[t for t in arange(0,T_s,1)]
    plt.plot(x,y)
    plt.grid()
    def w(a,b):    
        f = lambda t :(1/a**0.5)*exp(-0.5*((t-b)/a)**2)* (((t-b)/a)**2-1)*S(t)
        r= quad(f, -N, N)
        return round(r[0],3)
    x = arange(1,N_t,1)
    y = arange(1,N_t,1)
    z = array([w(i,j) for j in y for i in x])
    X, Y = meshgrid(x, y)
    Z = z.reshape(N_t-1 ,N_t-1)
    fig = plt.figure("–í–µ–π–≤–ª–µ—Ç - —Å–ø–µ–∫—Ç—Ä: –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–≥–æ –∫–æ–ª–µ–±–∞–Ω–∏—è")
    ax = Axes3D(fig)
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet)
    ax.set_xlabel(' –ú–∞—Å—à—Ç–∞–±:a')
    ax.set_ylabel('–ó–∞–¥–µ—Ä–∂–∫–∞: b')
    ax.set_zlabel('–ê–º–ø–ª–∏—Ç—É–¥–∞ –í–ü: $ N_{ab}$')
    plt.figure("2D-–≥—Ä–∞—Ñ–∏–∫ –¥–ª—è z = w (a,b)")
    plt.title('–ü–ª–æ—Å–∫–æ—Å—Ç—å ab —Å —Ü–≤–µ—Ç–æ–≤—ã–º–∏ –æ–±–ª–∞—Å—Ç—è–º–∏ –í–ü', size=12)
    plt.contourf(X, Y, Z,T_s)
    plt.show()

slider_T2 = widgets.IntSlider(min=20, max=50, value=20, continuous_update=True, description='–ü–µ—Ä–∏–æ–¥ T')
display(slider_T2)
drop_duration4 = widgets.Dropdown ( options=[1, 2, 5], value=2, description='–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å:')
display(drop_duration4)
drop_N = widgets.Dropdown ( options=[32, 64, 128,256,512], value=256, description='–ó–Ω–∞—á–µ–Ω–∏–µ N:')
display(drop_N)

button = widgets.Button(description="–ü–æ—Å—Ç—Ä–æ–∏—Ç—å")
button.on_click(on_button_clicked)
display(button)
```

**–°—É–º–º–∞ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–∏—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π**

–°–∏–≥–Ω–∞–ª –∏–º–µ–µ—Ç –≤–∏–¥ : $s(t):=A_{1}sin(\omega _{1}t)+A_{2}sin(\omega _{2}t)$

–≥–¥–µ: $A_{1}=A_{2}=1V,\omega_{1}=\frac{2\pi}{T_{1}},\omega_{2}=\frac{2\pi }{T_{2}},T_{2}=5T_1, T_3=7T_1, T_4=2T_1$.

$MHAT(t):=\frac{d^{2}}{dt^{2}}\left [ e^{-t^{2/2}} \right ]$, N:=256,
$\psi (a,b,t):=MHAT\left ( \frac{t-b}{a} \right ), W(a,b):=\int_{-N}^{N}\psi (a,b,t)f(t)dt$, a:=1...30, b:=0...50, .

–ù–∞ –ø–æ—Å–ª–µ–¥–Ω–µ–º –≥—Ä–∞—Ñ–∏–∫–µ –±—É–¥—É—Ç –ø—Ä–∏–≤–µ–¥–µ–Ω—ã —Å–µ—á–µ–Ω–∏—è –≤–µ–π–≤–ª–µ—Ç-—Å–ø–µ–∫—Ç—Ä–∞ –¥–ª—è –¥–≤—É—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –∞. –ü—Ä–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –Ω–µ–±–æ–ª—å—à–æ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–µ –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –º–∞—Å—à—Ç–∞–±–∞ a, –ø—Ä–∏ $a_{1}=2(a_{1}:1/\omega_{2})$, —Å–µ—á–µ–Ω–∏–µ —Å–ø–µ–∫—Ç—Ä–∞ –Ω–µ—Å–µ—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Ç–æ–ª—å–∫–æ –æ –≤—ã—Å–æ–∫–æ—á–∞—Å—Ç–æ—Ç–Ω–æ–π —Å–æ—Å—Ç–∞–≤–ª—è—é—â–µ–π —Å–∏–≥–Ω–∞–ª–∞, –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤—ã–≤–∞—è (–ø–æ–¥–∞–≤–ª—è—è) –µ–≥–æ –Ω–∏–∑–∫–æ—á–∞—Å—Ç–æ—Ç–Ω—ã–π –∫–æ–º–ø–æ–Ω–µ–Ω—Ç.


```python
from scipy.integrate import quad
from numpy import*
import matplotlib.pyplot as plt
from matplotlib import cm

def on_button_clicked(b):
    N=drop_N.value
    T=slider_T1.value
    T_s = T*drop_duration.value
    N_t=round(T*drop_duration.value/2)
    def S(t):
        if slider_num.value==2:
            return sin(2*pi*t/T)+sin(2*pi*t/(T*5))
        if slider_num.value==3:
            return sin(2*pi*t/T)+sin(2*pi*t/(T*5))+sin(2*pi*t/(T*7))
        if slider_num.value==4:
            return sin(2*pi*t/T)+sin(2*pi*t/(T*5))+sin(2*pi*t/(T*2))+sin(2*pi*t/(T*7))
    plt.figure(' –°—É–º–º–∞ {} –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–∏—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π'.format(slider_num.value))
    plt.title(' –°—É–º–º–∞ {} –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–∏—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π'.format(slider_num.value), size=12)
    y=[S(t) for t in arange(0,T_s,1)]
    x=[t for t in arange(0,T_s,1)]
    plt.plot(x,y)
    plt.grid()
    def w(a,b):    
        f = lambda t : (1/a**0.5)*exp(-0.5*((t-b)/a)**2)* (((t-b)/a)**2-1)*S(t)
        r= quad(f, -N, N)
        return round(r[0],3)
    x = arange(1,50,1)
    y = arange(1,50,1)
    z = array([w(i,j) for j in y for i in x])
    X, Y = meshgrid(x, y)
    Z = z.reshape(49, 49)
    plt.figure("2D-–≥—Ä–∞—Ñ–∏–∫ –¥–ª—è z = w (a,b)")
    plt.title('–ü–ª–æ—Å–∫–æ—Å—Ç—å ab —Å —Ü–≤–µ—Ç–æ–≤—ã–º–∏ –æ–±–ª–∞—Å—Ç—è–º–∏ –í–ü', size=12)
    plt.contourf(X, Y, Z, 100)
    plt.figure()
    q=[w(2,i) for i in y]
    p=[i for i in y]
    plt.plot(p,q,label='w(2,b)')
    q=[w(15,i) for i in y]
    plt.plot(p,q,label='w(15,b)')
    q=[w(30,i) for i in y]
    plt.plot(p,q,label='w(30,b)')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

slider_num = widgets.IntSlider(min=2, max=4, continuous_update=True, description='–ß–∏—Å–ª–æ –≥–∞—Ä–º–æ–Ω–∏–∫')
display(slider_num)
slider_T1 = widgets.IntSlider(min=10, max=30, continuous_update=True, description='–ü–µ—Ä–∏–æ–¥ T')
display(slider_T1)
drop_duration = widgets.Dropdown ( options=[2, 10, 20 , 40], value=20, description='–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å:')
display(drop_duration)
drop_N = widgets.Dropdown ( options=[32, 64, 128,256,512], value=256, description='–ó–Ω–∞—á–µ–Ω–∏–µ N:')
display(drop_N)

button = widgets.Button(description="–ü–æ—Å—Ç—Ä–æ–∏—Ç—å")
button.on_click(on_button_clicked)
display(button)
```

$s(t):=\begin{vmatrix}U, if t_{0}\leq t\leq t_{0}+\tau \\ 0, otherwise \end{vmatrix}$

$MHAT (t):=\frac{d^{2}}{dt^{2}}exp\left ( \frac{-t^{2}}{2} \right )$

$\psi (a,b,t):=MHAT\left ( \frac{t-b}{a} \right ), W(a,b):=\int_{-N}^{N}\psi (a,b,t)f(t)dt$

```python
from scipy.integrate import quad
from numpy import*
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

def on_button_clicked(b):
    N=drop_N2.value
    def S(t):
        U=5;t0=20;tau=slider_T5.value
        if  t0<=t<=t0+tau:
            return U
        else:
            return 0
    plt.figure()
    plt.title('–ü—Ä—è–º–æ—É–≥–æ–ª—å–Ω—ã–π –∏–º–ø—É–ª—å—Å', size=12)
    y=[S(t) for t in arange(0,120,1)]
    x=[t for t in arange(0,120,1)]
    plt.plot(x,y)
    plt.grid()
    def w(a,b):    
        f = lambda t :(1/a**0.5)*exp(-0.5*((t-b)/a)**2)* (((t-b)/a)**2-1)*S(t)
        r= quad(f, -N, N)
        return round(r[0],3)
    x = arange(1,100,1)
    y = arange(1,100,1)
    z = array([w(i,j) for j in y for i in x])
    X, Y = meshgrid(x, y)
    Z = z.reshape(99, 99)
    fig = plt.figure("3D-–≥—Ä–∞—Ñ–∏–∫ –≤–µ–π–≤–ª–µ—Ç —Å–ø–µ–∫—Ç—Ä–æ–≥—Ä–∞–º–º—ã")
    ax = Axes3D(fig)
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet)
    ax.set_xlabel(' –ú–∞—Å—à—Ç–∞–±:a')
    ax.set_ylabel('–ó–∞–¥–µ—Ä–∂–∫–∞: b')
    ax.set_zlabel('–ê–º–ø–ª–∏—Ç—É–¥–∞ –í–ü: $ N_{ab}$')
    plt.figure("2D-–≥—Ä–∞—Ñ–∏–∫ –¥–ª—è z = w (a,b)")
    plt.title('–ü–ª–æ—Å–∫–æ—Å—Ç—å ab —Å —Ü–≤–µ—Ç–æ–≤—ã–º–∏ –æ–±–ª–∞—Å—Ç—è–º–∏ –í–ü', size=12)
    plt.contourf(X, Y, Z,100)
    plt.show()

slider_T5 = widgets.IntSlider(min=10, max=60, value=30, continuous_update=True, description='–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å')
display(slider_T5)
drop_N2 = widgets.Dropdown ( options=[32, 64, 128,256,512], value=256, description='–ó–Ω–∞—á–µ–Ω–∏–µ N:')
display(drop_N)

button = widgets.Button(description="–ü–æ—Å—Ç—Ä–æ–∏—Ç—å")
button.on_click(on_button_clicked)
display(button)
```
